```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
df = pd.read_csv(r"C:\Users\siris\Downloads\data.csv.zip")
df.head(5)
```

```python
df.dropna()
```

```python
lemmatizer = WordNetLemmatizer()
import re
sentences = []
for i in range(len(df)):
    words = re.sub('[^a-zA-Z]',' ',df.Headline[i])
    words = nltk.word_tokenize(words)
    words = [lemmatizer.lemmatize(words[i]) for i in range(len(words)) if not words[i] in stopwords.words('english')]
    words=' '.join(words)
    sentences.append(words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_id = TfidfVectorizer(max_features=100000,ngram_range=(1,3))
data = tf_id.fit_transform(sentences).toarray()
print(tf_id.get_feature_names())
```

```python
from sklearn.model_selection import train_test_split
df.dropna()
Y=df['Label']
X_train,X_test,Y_train,Y_test = train_test_split(data,Y,test_size=0.3)
```

```python
data = pd.DataFrame(X_train,columns=tf_id.get_feature_names())
data
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
model = PassiveAggressiveClassifier()
model.fit(X_train,Y_train)
```

```python
Y_predict = model.predict(X_test);
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_predict)
```
